```python
import matplotlib.pyplot as plt
import numpy as np

# Players and their salary data (simplified example)
Players = ["Dhoni", "Kohli", "Sachin", "Rahul"]
latest_salaries = [18995624, 18862875, 23500000, 23180790]  # Salaries for Season 2024

# Create Bar Chart
plt.figure(figsize=(5, 5))

bars = plt.bar(
    Players, 
    latest_salaries, 
    color=['blue', 'green', 'orange', 'red'],
    label="Player Salary"
)

# Add Title and Axis Labels
plt.title("IPL Players Salary Comparison - Season 2024", fontsize=14, color="blue")
plt.xlabel("Players", fontsize=12)
plt.ylabel("Salary (in ₹)", fontsize=12)

# Add Legend (single legend for all bars)
plt.legend(["Player Salary"])

# Add Salary Values on Top of Each Bar (Labeling Bars)
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval + 50000, f'{yval:,}', 
             ha='center', va='bottom', fontsize=7)

# Show the plot
plt.show()
```
